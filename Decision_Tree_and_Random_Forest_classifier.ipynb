```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/breast_cancer.csv")
```

```python
#splitting the data
```

```python
X = df.iloc[:,:-1]                                   #we can use df.drop or this to drop the target variable which is outcome
X.head()
```

```python
y = df.outcome
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#Decision Tree Classifier
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model1 = DecisionTreeClassifier(max_depth=4,splitter='best',criterion='entropy')
```

```python
model1.fit(x_train,y_train)
```

```python
d_y_predict = model1.predict(x_test)
```

```python
d_y_predict
```

```python
d_y_predict_train = model1.predict(x_train)
```

```python
#Evaluating DTC
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print('Testing Accuracy = ', accuracy_score(y_test,d_y_predict))
print('Training Accuracy = ', accuracy_score(y_train,d_y_predict_train))
```

```python
pd.crosstab(y_test,d_y_predict)
```

In the cell at the intersection of '0' (actual class) and '0' (predicted class), you have 87. This means that 87 instances in your test data had a true class of '0' and were correctly predicted as '0' by the model.

In the cell at the intersection of '0' (actual class) and '1' (predicted class), you have 3. This means that 3 instances in your test data had a true class of '0' but were incorrectly predicted as '1' by the model.

In the cell at the intersection of '1' (actual class) and '0' (predicted class), you have 2. This means that 2 instances in your test data had a true class of '1' but were incorrectly predicted as '0' by the model.

In the cell at the intersection of '1' (actual class) and '1' (predicted class), you have 51. This means that 51 instances in your test data had a true class of '1' and were correctly predicted as '1' by the model.

```python
print(classification_report(y_test,d_y_predict))
```
