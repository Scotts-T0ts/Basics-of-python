```python
#Outlier removal by replacement with median
# removal methods- IQR(Inter Quartile Range),Z-score,Percentile method
```

```python
q1 = df.bmi.quantile(0.25) #Q1
q3 = df.bmi.quantile(0.75) #Q3
```

```python
print(q1)
print(q3)
```

```python
IQR = q3-q1
```

```python
IQR
```

```python
upper_limit = q3+1.5*IQR                 #upper limit of box plot
upper_limit
```

```python
lower_limit =q1-1.5*IQR
lower_limit
```

```python
df.median()                              #replacing with median
```

```python
df['bmi'] = np.where(df['bmi']>upper_limit,30,df['bmi'])
```

```python
sns.boxplot(df.bmi)
```

```python
#Outlier removal using IQR
```

```python
sns.boxplot(df.bmi)
```

```python
from scipy import stats
```

```python
bmi_zscore = stats.zscore(df.bmi)
bmi_zscore
```

```python
df_z = df[np.abs(bmi_zscore)<=3]
```
