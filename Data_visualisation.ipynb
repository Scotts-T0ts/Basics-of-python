```python
days=['Sun','Mon','Tue','Wed','Thu','Fri','Sat']
steps=[6745,8765,7456,4567,8799,3456,11567]
lwsteps=[8745,6784,11457,5964,7845,7945,13457]
plt.plot(days,steps,'o-g')
plt.plot(days,lwsteps,'o-m')
plt.title("Steps Count - Last Week & This Week")                         #Title
plt.xlabel('Days Of the week')                                           #X-axis label
plt.ylabel('Steps Walked')                                               #Y-axis label
plt.grid(True)
plt.legend(['Current Week','Last Week'])                                  #label
```

```python
plt.figure(figsize = (10,10))
age = [32,34,41,32,21]
name=['VK','Rohit','MSD','HP','SG']
plt.bar(name,age)
```

```python
name=['VK','Rohit','MSD','HP','SG']
ot=[29,48,32,21,41]
ct=[21,32,21,10,30]
plt.bar(name,ot,label='Open Tickets',width=0.5,color='k')
plt.bar(name,ct,label='Closed Tickets',width=0.5,color='b')
plt.xlabel('Name->')
plt.ylabel('Tickets->')
plt.legend()
```

```python
import matplotlib.pyplot as plt

age = [32, 34, 41, 32, 21]
name = ['VK', 'Rohit', 'MSD', 'HP', 'SG']
colors = ['g', 'b', 'r', 'k', 'y']

plt.pie(age, labels=name, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Age Distribution")

plt.show()
```
